This notebook can be used to visualize functions that we want to find maximum of.
This is part of https://github.com/alphatamago/opt

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
# One can change this to a different 2d function
f = lambda x, y: - (x**2 + y**2)
```

```python
def visualize_2dfunc(f):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    X, Y, Z = axes3d.get_test_data(0.05)
    Z = f(X, Y)
    ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)
    plt.show()
```

```python
visualize_2dfunc(f)
```

```python
f(*[0, 0])
```

```python
visualize_2dfunc(lambda x, y: 3 * x**2)
```
